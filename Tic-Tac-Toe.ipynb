```python
from random import shuffle
from IPython.display import clear_output
players = ["Player 1","Player 2"]
def choose_start_player():
    shuffle(players)
    if players[0] == "Player 1":
        print(f"{players[0]} will start with X.")
    else:
        print(f"{players[0]} will start with O.")
    return players

row1 = [" ", " ", " "]
row2 = [" ", " ", " "]
row3 = [" ", " ", " "]
def display_board():
    i = 0
    while i<3:
        if i == 1:
            print("    "+"    |    ".join(row1))
            i+=1
        else:
            print("\t | \t   |")
            i+=1
    i = 0
    print("------------------------------")
    while i<3:
        if i == 1:
            print("    "+"    |    ".join(row2))
            i+=1
        else:    
            print("\t | \t   |")
            i+=1
    i = 0
    print("------------------------------")
    while i<3:
        if i == 1:
            print("    "+"    |    ".join(row3))
            i+=1
        else:    
            print("\t | \t   |")
            i+=1 
        
d = {'7':row1[0], '8':row1[1], '9':row1[2], '4':row2[0], '5':row2[1], '6':row2[2], '1':row3[0], '2':row3[1], '3':row3[2]}

def choose_position():
    lst = ['1','2','3','4','5','6','7','8','9']
    position = "wrong"
    while position not in lst:
        position = input("Please choose an integer (1-9): ")
        if position not in lst:
            print("Sorry, invalid choice.")
            clear_output()
    return position

def assign_position(pos):
    choice = "wrong"
    lst = ['O','X']
    while choice not in lst:
        choice = input("Please enter your choice: ")
        if choice not in lst:
            print("Sorry, invalid choice.")
        clear_output()
    if pos in ['7','8','9']:
        row1[int(pos)-7] = d[pos] = choice
    elif pos in ['4','5','6']:
        row2[int(pos)-4] = d[pos] = choice
    else:
        row3[int(pos)-1] = d[pos] = choice
    
def display_dict():
    for pos,val in d.items():
        print(f"{pos} - {val}")
    print()
    print(row1)
    print(row2)
    print(row3)
        
def game_on():
    choice = "wrong"
    while choice not in ['Y','N']:
        choice = input("Do you want to keep playing (Y/N): ")
        if choice not in ['Y','N']:
            print("Sorry, invalid choice")
        return choice
    
def game_logic():
    space = ' '
    if row1[0] == row2[0] == row3[0] == 'X' or row1[1] == row2[1] == row3[1] == 'X' or row1[2] == row2[2] == row3[2] == 'X':
        return "Player 1 Wins!"
    elif row1[0] == row2[0] == row3[0] == 'O' or row1[1] == row2[1] == row3[1] == 'O' or row1[2] == row2[2] == row3[2] == 'O':
        return "Player 2 Wins!"
    elif row1[0] == row1[1] == row1[2] == 'X' or row2[0] == row2[1] == row2[2] == 'X' or row3[0] == row3[1] == row3[2] == 'X':
        return "Player 1 Wins!"
    elif row1[0] == row1[1] == row1[2] == 'O' or row2[0] == row2[1] == row2[2] == 'O' or row3[0] == row3[1] == row3[2] == 'O':
        return "Player 2 Wins!"
    elif row1[0] == row2[1] == row3[2] == 'X':
        return "Player 1 Wins!"
    elif row1[0] == row2[1] == row3[2] == 'O':
        return "Player 2 Wins!"
    elif row1[2] == row2[1] == row3[0] == 'X':
        return "Player 1 Wins!"
    elif row1[2] == row2[1] == row3[0] == 'O':
        return "Player 2 Wins!"
    elif space not in row1 and space not in row2 and space not in row3:
        return "It's a Draw!"
    else:
        pass
    
def result(winner):
    if winner in ["Player 1 Wins!","Player 2 Wins!"]:
        print(winner)
        return "----------GAME OVER----------"
    elif winner == "It's a Draw!":
        print(winner)
        return "----------GAME OVER----------"
    else:
        pass

choose_start_player()
done = True
while done == True:
    display_board()
    position = choose_position()
    assign_position(position)
    winner = game_logic()
    #display_dict()
    word = result(winner)
    if word == "----------GAME OVER----------":
        print(word)
        choice = game_on()
        if choice == 'Y':
            row1 = [" ", " ", " "]
            row2 = [" ", " ", " "]
            row3 = [" ", " ", " "]
            done = True
        else:
            done = False
    else:
        continue
```
